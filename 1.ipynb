**1. Створення  Xls документу, та отримання даних.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("data.xlsx")
df
```

**2. Візуалізація залежностей.**

```python
fig, ax = plt.subplots(1, 3, figsize = (12, 4))
fig.tight_layout(h_pad = 6)
fig.suptitle("Dependencies of properties", fontsize = 16)
plt.subplots_adjust(top = 0.75)

df = df.sort_values(by=['price'])
ax[0].set(title = "price / consumption", xlabel = "price, red.thousands ", ylabel = "consumption")
ax[0].plot(df["price"]/1000, df["consumption"], color = '#607c8e')

df = df.sort_values(by=['consumption'])
ax[1].set(title = "consuption / speed", xlabel = "consuption", ylabel = "speed")
ax[1].scatter(df["consumption"], df["speed"], color = '#607c8e')

df = df.sort_values(by=['payload'])
ax[2].set(title = "payload / speed", xlabel = "payload", ylabel = "speed")
ax[2].scatter(df["payload"], df["speed"], color = '#607c8e')



plt.show()
```

**3. Побудова гістограми.**

```python
plt.hist(df['speed'], 5, color = '#607c8e', width = 26)
plt.title("Histogram of speeds")
plt.xlabel("Speeds")
plt.ylabel("Number of brands")
plt.grid(axis='y', alpha=0.5)
plt.grid(axis='x', alpha=0.5)
plt.show()
```

**4. Побудова коробкового графіку.**

```python
sns.boxplot(x=df['speed'], data=df, color = '#607c8e').set_title("Speed distribution of marks of Toyota")
plt.show()

```

**5. Побудова парного графіку.**

```python
sns.pairplot(df, height=2.5)
plt.show()
```
